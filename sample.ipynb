```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
dogs_wiskers_length = np.random.normal(loc=5,scale=1,size=10)
dogs_ear_flappiness_index = np.random.normal(loc=8,scale=1,size=10)
```

```python
cats_wiskers_length = np.random.normal(loc=8,scale=1,size=10)
cats_ear_flappiness_index = np.random.normal(loc=5,scale=1,size=10)
```

```python
# Combine data
```

```python
plt.scatter(dogs_wiskers_length, dogs_ear_flappiness_index, color='blue', label='Dogs')
plt.scatter(cats_wiskers_length, cats_ear_flappiness_index, color='orange', label='Cats')
plt.xlabel('Wiskers Length (cm)')
plt.ylabel('Ear Flappiness Index')
plt.title('Dogs vs Cats: Wiskers Length vs Ear Flappiness')
plt.legend()
plt.grid(True)
plt.show()
```

```python
 #Implement random linear classifier algorithm
def random_linear_classifier(data_dogs,data_cats,k,d):
#d is the number of features
 best_error= float('inf')
 best_theta=None
 best_theta0=None

 for i in range(k):
  theta = np.random.normal(size=d)
  theta0 = np.random.normal()

  error = computer_error(data_dogs,data_cats,theta,theta0)

  if error < best_error:
     best_error = error
     best_theta = theta
     best_theta0 = theta0

 return best_theta, best_theta0
```

```python
def computer_error(data_dogs,data_cats,theta,theta0):
  error = 0

  for x in data_dogs:
     if np.dot(theta,x) + theta0 <= 0:
        error += 1

  for x in data_cats:
     if np.dot(theta,x) + theta0 > 0:
        error += 1

  return error
```

```python
# prepare data for algorithm
dogs_data = np.vstack((dogs_wiskers_length,dogs_ear_flappiness_index)).T
cats_data = np.vstack((cats_wiskers_length,cats_ear_flappiness_index)).T
```

```python

dogs_data
```

```python
# Run random linear classifier algorithm
k=100
d=2
best_theta, best_theta0 = random_linear_classifier(dogs_data,cats_data,k,d)
```

```python
# plot decision boundary
x_vals = np.linspace(2, 10, 100)
y_vals = (-best_theta[0] /best_theta[1] ) * x_vals - (best_theta0 / best_theta[1])
```

```python
plt.scatter(dogs_wiskers_length, dogs_ear_flappiness_index, color='blue', label='Dogs')
plt.scatter(cats_wiskers_length, cats_ear_flappiness_index, color='orange', label='Cats')
plt.plot(x_vals, y_vals, color='green', label='Decision Boundary')
plt.xlabel('Wiskers Length (cm)')
plt.ylabel('Ear Flappiness Index')
plt.title('Dogs vs Cats: Wiskers Length vs Ear Flappiness')
plt.legend()
plt.grid(True)
plt.show() 
```
